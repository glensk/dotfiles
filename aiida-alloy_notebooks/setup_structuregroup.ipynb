```python
def retrieve_allstrcture_fromgroup(group_name, structure_filters={}):
    qb = QueryBuilder()
    
    group_filters={"name":group_name}
    qb.append(Group,
              filters=group_filters,
              tag='selected_group'
             )
    if len(qb.all()) != 1:
        raise ValueError("There are no groups with group_name: {}".format(group_name))
    
    qb.append(StructureData, 
              filters=structure_filters,
              member_of='selected_group')
            
    return [x[0] for x in qb.all()]
```
