```python
%aiida
import ase
import ase.io
import random

import numpy as np
import matplotlib.pyplot as plt
import matplotlib

from aiida.orm.group import 
from aiida.orm.group import Group

matplotlib.rcParams['figure.dpi'] = 800
```

```python
db_path="/home/daniel/RnD/DataSets/Al6xxxDB_fps-50000/fps-5000.data.comments_replaced"
group_name = "Al6xxxDB"
group_description = "AiiDA group for a 5000 structure subset of the Al6xxxDB"
```

```python
# Setup/Retrieve the Group
g = Group.get_or_create(name=group_name, description=group_description)[0]

# Loop over structures in the db_path, storing the nodes then adding them to group
i = 0
while True:
    print "Storing structure index: {}".format(i)
    try: 
        ase_structure = ase.io.read(db_path, index=i, format="runner")
    except StopIteration:
        break
        
    # setup and store the ase structure as an aiida StructureData node
    StructureData = DataFactory("structure")
    aiida_structure = StructureData()
    aiida_structure.set_ase(ase_structure)
    aiida_structure_stored = aiida_structure.store()

    # add in the comment line if possible 
    try:
        comment = ase_structure.comment.strip().split()[-1][3:]
        aiida_structure_stored.set_extra("comment",comment)
    except AttributeError:
        pass 

    # add the structure to the group
    g.add_nodes(aiida_structure_stored)
    g.store()
        
    i += 1
```
